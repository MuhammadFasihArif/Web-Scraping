```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://books.toscrape.com/'
```

```python
response = requests.get(url)
if response.status_code == 200:
    print('request successful')
else:
    print('request failed')
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
books = soup.find_all('h3')
count=0
for book in books:
    book_url= book.find('a')['href']
    book_response = requests.get(url+book_url)
    book_soup = BeautifulSoup(book_response.content, 'html.parser')

    title = book_soup.find('h1').text.strip()
    price = book_soup.find('p', class_='price_color').text.strip()
    stock = book_soup.find('p', class_='instock availability').text.strip()
    table = book_soup.find('table', class_='table table-striped')
    upc = None
    for row in table.find_all('tr'):
        if row.find('th').text == 'UPC':
            upc = row.find('td').text
            break

    # Since 'star-rating' class doesn't contain text, extract it differently
    review = book_soup.find('p', class_='star-rating')['class'][1]
    count = count+1
    print(count,title, price, stock, upc, review)
```

```python
books_data = []
for pages in range(1,36):
    url = f'https://books.toscrape.com/catalogue/page-{pages}.html'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    books = soup.find_all('h3')
    count=0
    for book in books:
        book_url= book.find('a')['href']
        # Construct the full book URL correctly
        book_response = requests.get('https://books.toscrape.com/catalogue/' + book_url)
        book_soup = BeautifulSoup(book_response.content, 'html.parser')

        title = book_soup.find('h1').text.strip()
        price = book_soup.find('p', class_='price_color').text.strip()
        # Handle the case where stock information might be missing
        stock_element = book_soup.find('p', class_='instock availability')
        stock = stock_element.text.strip() if stock_element else "Not available"
        table = book_soup.find('table', class_='table table-striped')
        upc = None
        for row in table.find_all('tr'):
            if row.find('th').text == 'UPC':
                upc = row.find('td').text
                break

        # Since 'star-rating' class doesn't contain text, extract it differently
        review = book_soup.find('p', class_='star-rating')['class'][1]
        count = count+1
        print(count,title, price, stock, upc, review)
        books_data.append({
            'title': title,
            'price': price,
            'stock': stock,
            'upc': upc,
            'rating': review
        })
df = pd.DataFrame(books_data)
df.to_csv('books_data.csv', index=False)

print("Scraping completed and data saved to books_data.csv")
```
